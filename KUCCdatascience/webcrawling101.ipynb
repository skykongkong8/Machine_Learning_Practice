```python
html = BeautifulSoup(raw.text, "html.parser")
clips = html.select("div.inner")
title = clips[0].select_one("dt.title")
chn = clips[0].select_one("dd.chn")
hit = clips[0].select_one("span.hit")
like = clips[0].select_one("span.like")
print(title.text)
print(chn.text)
print(hit.text)
print(like.text)
```

```python
#이걸 전체에 적용하고싶다면 for문을 이용해서 clips를 반복문으로 만들어주면 된다.
for clip in clips:
    title = clip.select_one("dt.title")
    chn = clip.select_one("dd.chn")
    hit = clip.select_one("span.hit")
    like = clip.select_one("span.like")
    
    print(title.text.strip())
    print(chn.text.strip())
    print(hit.text.strip())
    print(like.text.strip())
```

```python
#다음 페이지로 넘어가면서 해보기
MAX=100
sum = 0
for n in range(1,MAX,10):
    raw = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=머신러닝"+str(n),headers={'User-Agent':'Mozilla/5.0'})
    html = BeautifulSoup(raw.text, "html.parser")
    articles = html.select("ul.type01 > li")
    sum+=1
    for article in articles:
        title = article.select_one("a._sp_each_title").text
        source = article.select_one("span._sp_each_source").text

        print(title,source,sum)
```

```python
f = open("navertv.csv",'w')
f.write('제목','채널명','재생 수','좋아요 수\n')

raw = requests.get("http://tv.naver.com/r")
html = BeautifulSoup(raw.text, "html.parser")
clips = html.select("div.cds") #class는.이고 아이디는#이다
for clip in clips:
    title = clip.select_one("dt.title").text.strip()
    chn = clip.select_one(".chn").text.strip()
    hit = clip.select_one(".hit").text.strip()
    like = clip.select_one(".like").text.strip()
f.write(title+','+chn+','+hit+','+like+'\n')
f.close()
```
