```python
import pandas as pd
```

```python
df = pd.read_csv('data/train.csv')
df.head(10)
```

```python
df[['Sex', 'Age', 'SibSp', 'Parch']].head()
```

```python
df_test = pd.read_csv("data/test.csv")
df_test.head()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
#fillna 빠져있는 빈칸에 무언가를 넣어주는 함수-- 이 경우 평균값을 채워주게된다. !데이처가 구멍똟려있을때 대처방법!
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean())
df.head()
```

```python
df.loc[df['Age']<10, 'Age'] =0
df.loc[(df['Age']>=10) & (df['Age']<20), 'Age'] = 1
df.loc[(df['Age']>=20) & (df['Age']<30), 'Age'] = 2
df.loc[(df['Age']>=30) & (df['Age']<40), 'Age'] = 3
df.loc[(df['Age']>=40) & (df['Age']<50), 'Age'] = 4
df.loc[df['Age']>=50, 'Age'] =5

df_test.loc[df_test['Age']<10, 'Age'] =0
df_test.loc[(df_test['Age']>=10) & (df_test['Age']<20), 'Age'] = 1
df_test.loc[(df_test['Age']>=20) & (df_test['Age']<30), 'Age'] = 2
df_test.loc[(df_test['Age']>=30) & (df_test['Age']<40), 'Age'] = 3
df_test.loc[(df_test['Age']>=40) & (df_test['Age']<50), 'Age'] = 4
df_test.loc[df_test['Age']>=50, 'Age'] =5
#jupyter notebook > edit > find and replace를 통해 보다 쉽게 이 코드를 작성할 수 있다.
#이걸 두 번 실행하게 되면 모두 0이 되어버리니 조심할 것
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch']
df.head()
```

```python
train = df[['Survived', 'Sex', 'Age', "FamilySize"]]
test = df_test[['Sex', "Age", 'FamilySize']]
train.head()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
df.isnull().sum()
```

```python
df['Embarked'].value_counts()
#S가 가장 많은 값임을 확인한다
```

```python
df['Embarked'] = df['Embarked'].fillna("S")
#가장 많은 값으로 채우고 isnull 확인 -- embarked의 NaN 값이 채워졌음을 볼 수 있다
df.isnull().sum()
```

```python
train = df[['Survived', 'Sex', 'Age', "FamilySize", "Fare", "Embarked"]]
test = df_test[['Sex', "Age", 'FamilySize', "Fare", "Embarked"]]
train.head()
```

```python
#숫자로 바꾸는게 인식하기 훨씬 편하기 떄문에 'sex'와 'Embarked'를 숫자로 바꿔야한다.
df.loc[df['Sex']=='female', "Sex"] = 1
df.loc[df['Sex']=='male', "Sex"] = 0
df.loc[df['Embarked']=='S', "Embarked"] = 0
df.loc[df['Embarked']=='C', "Embarked"] = 1
df.loc[df['Embarked']=='Q', "Embarked"] = 2


df_test.loc[df_test['Sex']=='female', "Sex"] = 1
df_test.loc[df_test['Sex']=='male', "Sex"] = 0
df_test.loc[df_test['Embarked']=='S', "Embarked"] = 0
df_test.loc[df_test['Embarked']=='C', "Embarked"] = 1
df_test.loc[df_test['Embarked']=='Q', "Embarked"] = 2
```

```python
df.head()
```

```python
df_test.head()
```

```python
train.loc[df['Sex']=='female', "Sex"] = 1
train.loc[df['Sex']=='male', "Sex"] = 0
train.loc[df['Embarked']=='S', "Embarked"] = 0
train.loc[df['Embarked']=='C', "Embarked"] = 1
train.loc[df['Embarked']=='Q', "Embarked"] = 2


test.loc[df_test['Sex']=='female', "Sex"] = 1
test.loc[df_test['Sex']=='male', "Sex"] = 0
test.loc[df_test['Embarked']=='S', "Embarked"] = 0
test.loc[df_test['Embarked']=='C', "Embarked"] = 1
test.loc[df_test['Embarked']=='Q', "Embarked"] = 2
train.head()
```

```python
test.head()
```

```python
x_train = train[['Sex', 'Age', "FamilySize", "Fare", "Embarked"]]
y_train = train['Survived']
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(x_train, y_train)

print('Training set Acuuracy:', tree.score(x_train, y_train))
```

```python
prediction = tree.predict(x_test)
prediction
```
