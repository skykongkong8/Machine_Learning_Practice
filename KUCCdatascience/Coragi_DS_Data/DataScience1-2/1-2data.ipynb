```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train = pd.read_csv("data/train.csv")
train.head()
```

```python
sns.set()
```

```python
sns.relplot(data = train, x='Age', y = 'Survived') #1 relplot: 산포도--dataset을 가져오고, 각각의 축으 어떤 요소를 넣을 것인지
```

```python
sns.relplot(data = train, x = 'Age', y = 'Survived', hue = 'Sex', aspect = 0.5)#성별에 색을 부여하기, 사이즈를 조절하기(너무 겹쳐있는경우)
```

```python
sns.catplot(data=train, x = 'Pclass', y = 'Fare', aspect =4)
sns.catplot(data = train, x= 'Pclass', y = 'Fare', jitter = False, aspect = 4)
sns.catplot(data = train, x= 'Pclass', y = 'Fare', hue = 'Sex', aspect = 4)
```

```python
sns.pairplot(data = train, hue = 'Sex',
            x_vars = ['Pclass','Age','Fare'],
            y_vars = ['Survived', 'Age', 'Fare'], height = 3) #pairplot--여러 가지 형식의 그래프를 모두 출력--어떻게 탐색을 시작할지 방향을 잡아줌
```

```python
sns.catplot(data = train, x = 'Pclass', y = 'Fare', kind = 'box')
sns.catplot(data = train, x = 'Pclass', y = 'Fare', kind = 'box', hue = 'Sex') #kind를 추가해서 box형태롤 변경하기
```
