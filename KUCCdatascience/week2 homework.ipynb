```python
import requests
from bs4 import BeautifulSoup

for n in range(1, 4) :
    raw = requests.get("https://news.ycombinator.com/news?p="+str(n),
                       headers = {'User-Agent' : "Mozilla/5.0"})

    html = BeautifulSoup(raw.text, 'html.parser')

    articles = html.select("tr.athing")

    for article in articles :
        title = article.select_one("a.storylink").text
        rank = article.select_one("span.rank").text
        source = article.select_one("span.sitestr").text

        print('제목 : ' + title)
        print('순위 : ' + rank)
        print('출처 : ' + source)
```

```python
for n in range(1,6):
    raw = requests.get("https://series.naver.com/ebook/top100List.nhn?page="+str(n))
    html = BeautifulSoup(raw.text , "html.parser")
    books = html.select("#content li")
    for book in books:
        bookname = book.select_one("#content a strong").text.strip()
        author = book.select_one("span.writer").text.strip()
        
        print("제목 : "+bookname)
        print("저자 이름 : "+author)
```

```python
import requests
from bs4 import BeautifulSoup

f= open("naverebooktop100.csv","w")
f.write("제목, 저자이름\n")
for n in range(1,6):
    raw = requests.get("https://series.naver.com/ebook/top100List.nhn?page="+str(n))
    html = BeautifulSoup(raw.text , "html.parser")
    books = html.select("#content li")
    for book in books:
        bookname = book.select_one("#content a strong").text.strip()
        author = book.select_one("span.writer").text.strip()
        bookname.replace(",","")
        author.replace(",","")
        f.write(bookname+","+author+"\n")
f.close()
```
