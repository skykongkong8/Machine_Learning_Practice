```python
"""Review: diabetes_dataset with polynomial_regression"""
from sklearn import datasets
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

#우선, sklearn에 내장되어 있는 data로 연습해보자
diabetes_dataset = datasets.load_diabetes()

 #2차항까지 있는 가설함수를 쓰기로 하자
polynomial_transformer = PolynomialFeatures(2)

#통 dataset에서 data와 feature만 분리!
polynomial_data = polynomial_transformer.fit_transform(diabetes_dataset.data)
polynomial_feature_names = polynomial_transformer.get_feature_names(diabetes_dataset.feature_names)

diabetes_dataset.DESCR
```

```python

#X부분 df로 만들기--각 data와 column으로
X=pd.DataFrame(polynomial_data, columns=polynomial_feature_names)
#y부분(목표변수)
y = pd.DataFrame(diabetes_dataset.target, columns = ['diabetes'])
#train 부분과 test 부분으로 나누기
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=5)

#학습-기본 최적화 방법은 선형회귀 사용
model = LinearRegression()
model.fit(X_train, y_train)

y_test_predict = model.predict(X_test)
mse = mean_squared_error(y_test, y_test_predict)
mse**0.5

score=model.score(X_test, y_test)
score
```

```python
"""Review: Wine_sorting with Logistic_Regression"""
from sklearn import datasets
from sklearn.model_selection import train_test_split
#선형회귀가 아닌 로지스틱회귀(분류)
from sklearn.linear_model import LogisticRegression
import pandas as pd
wine_data = datasets.load_wine()
wine_data.feature_names, wine_data.target
```

```python
X = pd.DataFrame(wine_data.data, columns = wine_data.feature_names)
y=pd.DataFrame(wine_data.target, columns=['Y/N'])

X_train,X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=5)
y_train = y_train.values.ravel()
model = LogisticRegression(solver='saga', max_iter = 7500)
model.fit(X_train, y_train)
y_test_predict = model.predict(X_test)
score = model.score(X_test, y_test)

score
```

```python
"""Review: Lasso 모델과 L1 정규화"""
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from math import sqrt
import numpy as np
import pandas as pd

file_path = 'C:/Users/공성식/Documents/git_practice/machine_learning_study/csv_folder/insurance.csv'

insurance_df = pd.read_csv(file_path)
insurance_df = pd.get_dummies(data = insurance_df, columns = ['sex', 'smoker','region']) #one-hot encoding으로 숫자로 변환

X = insurance_df.drop(['charges'], axis =1)
polynomial_transformer = PolynomialFeatures(4)
polynomial_features = polynomial_transformer.fit_transform(X.values)
features = polynomial_transformer.get_feature_names(X.columns)

X=pd.DataFrame(polynomial_features, columns = features)
y = insurance_df[['charges']]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=5)

model = Lasso(alpha=1, max_iter=2000, normalize = True)
model.fit(X_train, y_train)
y_train_predict = model.predict(X_train)
y_test_predict = model.predict(X_test)

mse1 = mean_squared_error(y_train, y_train_predict)
mse2 = mean_squared_error(y_test, y_test_predict)
score = model.score(X_test, y_test)

mse1, mse2, score
```
